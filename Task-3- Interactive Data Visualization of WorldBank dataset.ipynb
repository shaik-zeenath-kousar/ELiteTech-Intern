```python
import pandas as pd
import plotly.express as px
from google.colab import files
import io
```

```python
uploaded = files.upload()
filename = next(iter(uploaded))
```

```python
df = pd.read_csv(io.StringIO(uploaded[filename].decode('utf-8')))
print("\nPreview of the uploaded data:")
display(df.head())
```

```python
expected_columns = ["Country", "Year", "GDP per capita", "Life Expectancy"]
if not all(col in df.columns for col in expected_columns):
    raise ValueError("Error: CSV must contain 'Country', 'Year', 'GDP per capita', and 'Life Expectancy'")
unique_countries = df["Country"].unique()
print("\nAvailable countries in dataset:", unique_countries)
selected_country = input("Enter a country from the list above: ")
selected_country = "USA"
df_country = df[df["Country"] == selected_country]
```

```python
fig_gdp = px.line(df_country, x="Year", y="GDP per capita",
                  title=f"GDP per Capita Over Time ({selected_country})")
fig_gdp.show()
```

```python
fig_life = px.line(df_country, x="Year", y="Life Expectancy",
                   title=f"Life Expectancy Over Time ({selected_country})")
fig_life.show()
```
