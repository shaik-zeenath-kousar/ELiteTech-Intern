```python
file_path="diamonds.csv"
data=pd.read_csv(file_path)
print(data)
```

```python
print("First 5 rows of the dataset:")
print(data.head())
```

```python
print("\nDataset Info:")
print(data.info())
```

```python
print("\nMissing Values:")
print(data.isnull().sum())
```

```python
print("\nDescriptive Statistics:")
print(data.describe())
```

```python
# Debugging: Print the column names
print("Column Names in Dataset:")
print(data.columns)

# Strip any leading or trailing spaces from column names (if needed)
data.columns = data.columns.str.strip()

# Check the first few rows of the dataset
print("\nFirst Few Rows of the Dataset:")
print(data.head())

# Verify the existence of the 'Carat' column
if 'Carat' in data.columns:  # Note: 'Carat' matches your dataset's column name
    # Proceed with visualization if column exists
    plt.figure(figsize=(8, 5))
    sns.histplot(data['Carat'], kde=True, bins=30, color='blue')
    plt.title('Distribution of Carat')
    plt.xlabel('Carat')
    plt.ylabel('Frequency')
    plt.show()
else:
    print("\nError: 'Carat' column not found in the dataset.")
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Cut', data=data, palette='muted', hue='Cut', legend=False)
plt.title('Count of Diamond Cuts')
plt.xlabel('Cut')
plt.ylabel('Count')
plt.show()
```

```python
if 'Cut' in data.columns:
    plt.figure(figsize=(8, 5))
    sns.countplot(x='Cut', data=data,hue='Cut', palette='muted')
    plt.title('Count of Diamond Cuts')
    plt.xlabel('Cut')
    plt.ylabel('Count')
    plt.show()
else:
    print("Error: 'Cut' column not found in the dataset.")
```

```python
if 'Carat' in data.columns and 'Price' in data.columns:  # Match case-sensitive column names
    plt.figure(figsize=(8, 5))
    sns.scatterplot(x='Carat', y='Price', data=data, alpha=0.6)  # Use 'Carat' and 'Price'
    plt.title('Price vs Carat')
    plt.xlabel('Carat')
    plt.ylabel('Price')
    plt.show()
else:
    print("Error: Required columns 'Carat' or 'Price' not found in the dataset.")
```

```python
if 'Cut' in data.columns and 'Price' in data.columns:
    plt.figure(figsize=(8, 5))
    sns.boxplot(x='Cut', y='Price', data=data, hue='Cut',palette='coolwarm')
    plt.title('Price by Cut')
    plt.xlabel('Cut')
    plt.ylabel('Price')
    plt.show()
else:
    print("Error: Columns 'Cut' and/or 'Price' not found in dataset.")
```

```python
if 'Color' in data.columns and 'Price' in data.columns:
    plt.figure(figsize=(8, 5))
    sns.boxplot(x='Color', y='Price', data=data,hue='Cut', palette='viridis')
    plt.title('Price by Color')
    plt.xlabel('Color')
    plt.ylabel('Price')
    plt.show()
else:
    print("Error: Columns 'Color' and/or 'Price' not found in dataset.")
```

```python
plt.figure(figsize=(10, 6))
correlation = data.corr(numeric_only=True)  # Use numeric columns only
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
if 'Cut' in data.columns and 'Price' in data.columns and 'Color' in data.columns:
    plt.figure(figsize=(10, 6))
    sns.catplot(x='Cut', y='Price', hue='Color', kind='box', data=data, height=6, aspect=1.5)
    plt.title('Price by Cut and Color')
    plt.xlabel('Cut')
    plt.ylabel('Price')
    plt.show()
else:
    print("Error: Columns 'Cut', 'Price', and/or 'Color' not found in dataset.")
```

```python
key_features = ['Carat', 'Price', 'Depth', 'Table']
missing_features = [feature for feature in key_features if feature not in data.columns]
if not missing_features:
    sns.pairplot(data, vars=key_features, hue='Cut', diag_kind='kde', palette='husl')
    plt.suptitle('Pairplot of Key Features', y=1.02)
    plt.show()
else:
    print(f"Error: Missing key features {missing_features} for pairplot.")
```

```python
print("\n### Insights ###")
print("- Carat has a strong positive correlation with price.")
print("- Premium and Ideal cuts generally have higher prices compared to Fair and Good cuts.")
print("- The color and clarity also have significant effects on the price of diamonds.")
if 'Carat' in data.columns and 'Price' in data.columns:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='Carat', y='Price', data=data, alpha=0.6)
    plt.title('Price vs Carat')
    plt.xlabel('Carat')
    plt.ylabel('Price')
    plt.savefig('price_vs_carat.png')
    print("\nSaved 'Price vs Carat' scatterplot as an image.")
else:
    print("Error: Columns 'Carat' and/or 'Price' not found in dataset.")
```
